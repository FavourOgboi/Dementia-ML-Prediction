## Alzheimer's

A mental capacity decline severe enough to interfere with day-to-day functioning is referred to as dementia. One frequent instance is memory loss. Dementia is a group of symptoms associated with a decline in memory and other cognitive abilities that make daily tasks difficult. It is not a specific condition.


### Identification of Dementia

One test is insufficient to diagnose dementia in an individual. Alzheimer's disease and other types of dementia are identified by doctors by:

- A comprehensive assessment of medical history
- A clinical assessment - Laboratory investigations
- Seeing the distinct alterations in thought processes, day-to-day activities, and behavior that are indicative of each kind of dementia

With great certainty, physicians can make the diagnosis of dementia. But pinpointing the exact kind of dementia can be hard because different dementias can have similar symptoms and neurological abnormalities. A physician may occasionally identify dementia without identifying the specific kind. In these situations, seeking advice from an expert like a geropsychologist or neurologist might be essential.

## Understanding the Data

### Summary
This dataset consists of a longitudinal study involving 150 subjects aged between 60 and 96. Each subject was scanned during two or more visits, with at least one year between each visit, resulting in a total of 373 imaging sessions. For each subject, 3 or 4 individual T1-weighted MRI scans obtained in single scan sessions are included. The subjects are all right-handed and include both men and women. Among them:
- 72 subjects were classified as nondemented throughout the study.
- 64 subjects were classified as demented at their initial visits and remained so for subsequent scans, including 51 individuals with mild to moderate Alzheimer’s disease.
- 14 subjects were initially nondemented but were later classified as demented during subsequent visits.

### Variables and Their Meanings
The dataset includes several key variables:

- **Subject.ID**: Unique identifier for each subject.
- **MRI.ID**: Unique identifier for each MRI session.
- **Group**: Indicates the dementia status of the subject. It can be one of three categories:
  - **Converted**: Subjects who were initially nondemented but later diagnosed as demented.
  - **Demented**: Subjects diagnosed with dementia at the initial visit and remained so.
  - **Nondemented**: Subjects who were not diagnosed with dementia throughout the study.
- **Visit**: The number of the visit (e.g., first visit, second visit, etc.).
- **MR.Delay**: The delay in days between the baseline visit and the MRI scan.

### Demographic Information
- **M.F**: Gender of the subject (M for male, F for female).
- **Hand**: Handedness of the subject (dropped since all subjects are right-handed).
- **Age**: Age of the subject at the time of the scan.
- **EDUC**: Years of education completed by the subject.
- **SES**: Socioeconomic status of the subject as assessed by the Hollingshead Index of Social Position, categorized from 1 (highest status) to 5 (lowest status).

### Clinical Information
- **MMSE**: Mini-Mental State Examination score, ranging from 0 (worst) to 30 (best). Used to assess cognitive impairment.
- **CDR**: Clinical Dementia Rating, which scores the severity of dementia:
  - 0 = No dementia
  - 0.5 = Very mild dementia
  - 1 = Mild dementia
  - 2 = Moderate dementia

### Derived Anatomic Volumes
- **eTIV**: Estimated total intracranial volume in cubic millimeters (mm³).
- **nWBV**: Normalized whole-brain volume, expressed as a percentage of all voxels in the atlas-masked image that are labeled as gray or white matter by the automated tissue segmentation process.
- **ASF**: Atlas scaling factor (unitless). It is the computed scaling factor that transforms native-space brain and skull to the atlas target, representing the determinant of the transform matrix.

### Mini–Mental State Examination (MMSE)
The MMSE is a 30-point questionnaire widely used in clinical and research settings to measure cognitive impairment. It is used to screen for dementia, estimate the severity and progression of cognitive impairment, and monitor changes in an individual's cognitive function over time. It is not intended to provide a diagnosis on its own.

### MMSE Interpretations
- A score of 24 or higher (out of 30) indicates normal cognition.
- Scores below 24 indicate cognitive impairment, categorized as:
  - Severe (≤9 points)
  - Moderate (10–18 points)
  - Mild (19–23 points)

Scores may need adjustment based on educational attainment and age. A perfect score of 30 does not rule out dementia, while very low scores strongly correlate with dementia, though other mental disorders can also affect MMSE results. Physical issues, such as hearing or motor deficits, can also interfere with the test's interpretation.

### Clinical Dementia Rating (CDR)
The CDR is a 5-point scale used to evaluate six domains of cognitive and functional performance, particularly in Alzheimer's disease and related dementias:
- Memory
- Orientation
- Judgment & Problem Solving
- Community Affairs
- Home & Hobbies
- Personal Care

Ratings are derived from a semi-structured interview with the patient and a reliable informant, such as a family member. The CDR Scoring Table provides guidelines for making appropriate ratings. An overall CDR score, calculated using the CDR Scoring Algorithm, helps characterize and track the level of impairment:
- 0 = Normal
- 0.5 = Very Mild Dementia
- 1 = Mild Dementia
- 2 = Moderate Dementia

### Estimated Total Intracranial Volume (eTIV)
The estimated total intracranial volume (eTIV) measures the cranial cavity's volume, outlined by the supratentorial dura matter or cerebral contour when dura is not clearly detectable. eTIV is used in studies involving cerebral structure analysis under different imaging modalities (e.g., MRI, DTI, SPECT, Ultrasound, CT). Due to its consistency during aging, eTIV is a reliable tool for correcting head size variations in morphological brain studies. It is reported as a covariate, along with age and gender, in regression analyses investigating neurodegenerative brain disorders like Alzheimer's disease. eTIV also helps evaluate age-related structural changes, atrophy patterns in mild cognitive impairment (MCI) and Alzheimer's disease (AD), and structural abnormalities in conditions such as schizophrenia and epilepsy.
